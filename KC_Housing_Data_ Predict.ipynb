```python
df = df.drop(['id', 'date'], axis=1)
```

```python
df
```

```python
x = df.iloc[:,1:]
y = df.iloc[:,0:1]
X = x.values
Y = y.values
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)
```

```python
from sklearn.metrics import r2_score

print(f'R² score: {r2_score(Y, lin_reg.predict(X))*100}')
```

# Train with KNeighborsRegressor Algorithm

```python
from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor(n_neighbors=3)
```

```python
knn_model.fit(X, y)
```

```python
prd= knn_model.predict(X)
prd
```

```python
print(f'R² score: {r2_score(Y, knn_model.predict(X))*100}')
```

#  KNeighborsRegressor __ R² score: 74.76283611632562  

# Train  With DecisionTreeRegressor

```python
from sklearn import tree
```

```python
dtree = tree.DecisionTreeRegressor()
dtree.fit(X, y)
```

```python
print(f'R² score: {r2_score(Y, dtree.predict(X))*100}')
```

# DecisionTreeRegressor __ R² score: 99.91338290544213

```python
dtreeprd = dtree.predict(X)
dtreeprd
```

# Visualize The Data with matplotlib

```python
newdata2pr = dtreeprd = dtree.predict(X)
plt.scatter(Y,newdata2pr, color="b")
plt.plot(Y,Y, color="r")
```
